### Programming learning algorithms and tuning their parameters to the best via grid search k fold cross validation:

```python
def train_test_split(df, train_size=0.7):
    df = np.array(df)
    return df[:int(len(df)*train_size)], df[int(len(df)*train_size):]
```

```python
train, test = train_test_split(df,train_size=0.7)
X_train, Y_train = train[:,:-1], train[:,-1:]
X_test, Y_test = test[:,:-1], test[:,-1:]
```

```python
class BaseModel:
    def f1(self, Y_pred, Y_real):
        TP = np.sum(Y_pred * Y_real)
        FP = len(Y_pred[(Y_pred == 1) & (Y_real == 0)])
        FN = len(Y_pred[(Y_pred == 0) & (Y_real == 1)])
        if TP + FP == 0 or TP + FN == 0:
            return None
        presicion = TP / (TP + FP)
        recall = TP / (TP + FN)
        return  2 * presicion * recall / (recall + presicion)
```

```python
class KNeighborsClassifier(BaseModel):
    def __init__(self, n_neighbors, weights, p):
        self.n_neighbors, self.weights, self.p = n_neighbors, weights, p
    def fit(self, X, Y):
        self.X, self.Y = X, Y
    def predict(self, X):
        predictions = np.array([])
        if self.weights == 'uniform':
            for x in X:
                XY = np.hstack(
                    [
                        self.Y,
                        np.power(
                            np.sum(
                                np.power(x - self.X,self.p),
                                axis=1),
                                1/self.p).reshape(-1,1)])
                XY = XY[XY[:,1].argsort()][:self.n_neighbors]
                counts = np.array(np.unique(XY[:,0],return_counts=True)).T
                predictions = np.append(predictions,counts[counts[:,1].argsort()][-1,0])
        elif self.weights == 'distance':
            for x in X:
                XY = np.hstack(
                    [
                        self.Y,
                        np.power(
                            np.sum(
                                np.power(x - self.X,self.p),
                                axis=1),
                                1/self.p).reshape(-1,1)])
                XY = XY[XY[:,1].argsort()][:self.n_neighbors]
                XY[:,1] = 1/XY[:,1]
                label_scores = np.array([[item,np.sum(XY[XY[:,0] == item,1])] for item in np.unique(XY[:,0])])
                predictions = np.append(predictions,label_scores[np.argmax(label_scores[:,1]),0])
        return predictions
```

```python
class MultinomialNB(BaseModel):
    def __init__(self, alpha=1, fit_prior=True):
        self.alpha, self.fit_prior = alpha, fit_prior
    def fit(self, X, Y):
        self.table = np.hstack([X,Y])
        self.label_counts = np.unique(Y,return_counts=True)
        if self.fit_prior:
            self.prior_prob = {label: self.label_counts[1][index]/ len(Y) for index, label in enumerate(self.label_counts[0])}
        else:
            self.prior_prob = {label: 1 for label in self.label_counts[0]}
    def predict(self, X):
        predictions = np.array([])
        labels = self.label_counts[0]
        for row in X:
            pred_label = None
            pred_label_prob = 0
            for label in labels:
                label_masking = self.table[:,-1] == label
                label_count = len(self.table[(label_masking)])
                label_prob = prod(
                    [
                        (len(self.table[(self.table[:,index] == x) & (label_masking)]) + self.alpha)/
                                (label_count + self.alpha * len(np.unique(self.table[:,index])))
                                  for index,x in enumerate(row)]) * self.prior_prob[label]
                if label_prob > pred_label_prob:
                    pred_label = label
                    pred_label_prob = label_prob
            predictions = np.append(predictions, pred_label)
        return predictions
```

```python
class gridSearchCV:
    def __init__(self, estimator, param_grid, cv):
        self.estimator, self.cv, param_grid = estimator, cv, param_grid
        self.combs = list(product(*[param_grid[key] for key in param_grid]))
        self.entries = list(param_grid)
    def fit(self, data):
        best_comb = None
        best_comb_score = 0
        for comb in self.combs:
            est = self.estimator(*comb)
            comb_scores = []
            chunks = np.array_split(data, self.cv)
            for i in range(self.cv):
                train_data = np.vstack(chunks[i+1:] + chunks[:i])
                test_data = chunks[i]
                X_train, Y_train = train_data[:,:-1], train_data[:,-1:]
                X_test, Y_test = test_data[:,:-1], test_data[:,-1:]
                est.fit(X_train, Y_train)
                Y_predicted = est.predict(X_test)
                score = est.f1(Y_predicted, Y_test.reshape(-1))
                if score != None:
                    comb_scores.append(score)
            comb_score = sum(comb_scores) / len(comb_scores)
            if comb_score > best_comb_score:
                best_comb_score = comb_score
                best_comb = comb
        return self.estimator(*best_comb) ,best_comb, best_comb_score
```

```python
knn_params = {
    'n_neighbors': np.arange(1,16),
    'weights': ['uniform','distance'],
    'p': np.arange(1, 4)
}
knn, best_comb, best_comb_score = gridSearchCV(KNeighborsClassifier, knn_params, 10).fit(np.array(df))
print(best_comb)
print(best_comb_score)
knn.fit(X_train,Y_train)
Y_predicted = knn.predict(X_test)
knn.f1(Y_predicted, Y_test.reshape(-1))
```

### KNN params: only the first nearest neighbor and uniform weights for data points works best via Manhattan distance

```python
mnb_params = {
    'alpha': np.arange(1,3),
    'fit_prior': [True, False],
    }
mnb, best_comb, best_comb_score = gridSearchCV(MultinomialNB, mnb_params, 10).fit(np.array(df))
print(best_comb)
print(best_comb_score)
mnb.fit(X_train,Y_train)
Y_predicted = mnb.predict(X_test)
mnb.f1(Y_predicted, Y_test.reshape(-1))
```

### Multinomial Naive Bayes params: alpha = 1 with uniform prior probabilities works fine.

### Training errors:
